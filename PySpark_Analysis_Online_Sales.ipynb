```
#Locate folder from the Databricks library - these were imported into Databricks from local machine
%fs ls /databricks-datasets/online_retail/data-001/
```

```
# specify path to the specific csv to be analyzed
path = "/databricks-datasets/online_retail/data-001/data.csv"

# load as text
data = spark.read.csv(path) 

# show sample - top 20 instances
data.take(20)
```

```
# read in file using csv format
df = spark.read.load(path,
                    format='com.databricks.spark.csv', 
                    header='true',
                    inferSchema='true')

# show 20 rows
display(df) #in addition to the data table the output also has the inferred schema for df
```

```
# For this we'll need a few functions
display( # shows the results in a grid
   df 
    .select("Country") # chooses just the 1 column
    .distinct() # removes duplicates
    .orderBy("Country") # sorts results in ascending
)
```

```
df.select("Country").show() #to show the Country column
```

```
#Creating an Aggregation from existing columns and renaming the resulting column
display(
  df
    .select(df["InvoiceNo"],(df["UnitPrice"]*df["Quantity"]).alias("Sale"))
    .groupBy("InvoiceNo")
    .sum()
)
```

```
r1=df.select(df["Country"], df["Description"], (df["UnitPrice"]*df["Quantity"]).alias("Total"))
display(r1)
```

```
r1.write.saveAsTable("product_sales_by_Country")
```
